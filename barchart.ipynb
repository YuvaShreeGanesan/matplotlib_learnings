```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

company = ['GOOG','AMZN','MSFT','FB']
revenue = [90,80,85,95]
profit = [40,2,34,12]

ypos = np.arange(len(company))

plt.barh(ypos, revenue, height=0.4, label="Revenue")
plt.barh(ypos + 0.4, profit, height=0.4, label="Profit")

plt.yticks(ypos + 0.2, company)
plt.xlabel("Revenue / Profit")
plt.ylabel("US Tech Stocks")
plt.title("US Tech Stocks")
plt.legend()

plt.show()
```
