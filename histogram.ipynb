```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python

blood_sugar = [
    113, 85, 90, 150, 149,
    88, 93, 115, 135, 80,
    77, 82, 129
]

plt.hist(blood_sugar, bins=5)
plt.xlabel("Blood Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Distribution")
plt.show()
```

```python
labels = ['Normal', 'Prediabetic', 'Diabetic']
sizes = [6, 4, 3]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Blood Sugar Category Distribution")
plt.show()
```
